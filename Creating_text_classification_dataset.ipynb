```
import numpy as np
import pandas as pd
```

```
data= pd.read_json(r"News_Category_Dataset_v2.json",lines=True)
data=data.drop(columns=["authors","date","link"])
data.head(10)
```

```
data.shape
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
data['category'].value_counts().plot.bar()
plt.show()
```

```
data.groupby("category").size()
```

```
data.replace({"category":{
    
    "ARTS":"Arts, Culture and Religion",
    "ARTS & CULTURE":"Arts, Culture and Religion",
    "RELIGION":"Arts, Culture and Religion",
    "CULTURE & ARTS":"Arts, Culture and Religion",
    "BLACK VOICES":"Voices",
    "QUEER VOICES":"Voices",
    "LATINO VOICES":"Voices",
    "BUSINESS":"Business and money",
    "MONEY":"Business and money",
    "GREEN":"ENVIRONMENT",
    "HEALTHY LIVING":"Wellness and Healthy Living",
    "WELLNESS":"Wellness and Healthy Living",
    "THE WORLDPOST":"World news",
    "WORLD NEWS":"World news",
    "WORLDPOST":"World news",
    "SCIENCE":"Education, science and Tech",
    "TECH":"Education, science and Tech",
    "COLLEGE":"Education, science and Tech",
    "EDUCATION":"Education, science and Tech",
    "FOOD & DRINK":"Food",
    "TASTE":"Food",
    "STYLE":"STYLE & BEAUTY" }},inplace=True)



data.drop(data[data["category"] == "PARENTING" ].index, inplace = True)
data.drop(data[data["category"] == "COMEDY" ].index, inplace = True)
data.drop(data[data["category"] == "HOME & LIVING"  ].index, inplace = True)
data.drop(data[data["category"] == "PARENTS" ].index, inplace = True)
data.drop(data[data["category"] == "WEDDINGS" ].index, inplace = True)
data.drop(data[data["category"] == "WOMEN" ].index, inplace = True)
data.drop(data[data["category"] == "IMPACT"].index, inplace = True)
data.drop(data[data["category"] == "DIVORCE"].index, inplace = True)
data.drop(data[data["category"] == "MEDIA" ].index, inplace = True)
data.drop(data[data["category"] == "GOOD NEWS" ].index, inplace = True)
data.drop(data[data["category"] == "FIFTY" ].index, inplace = True)
data.drop(data[data["category"] == "WEIRD NEWS" ].index, inplace = True)
```

```
plt.figure(figsize=(15,10))
data['category'].value_counts().plot.bar()
plt.show()
```

```
data_cropped=data.groupby('category').apply(lambda s: s.sample(min(len(s), 10000)))
plt.figure(figsize=(15,10))
data_cropped['category'].value_counts().plot.bar()
plt.show()
```

```
data_cropped.to_csv('New_news_classification_dataset.csv', index=False) 
```
